# Model Predict Testing

```python
import pandas as pd
import tensorflow as tf
```

```python
df = pd.read_csv("user_data_test_cleaned.csv", sep=';')
```

```python
df.head()
```

```python
y_test = df['is_scammer']
```

```python
x_test = df.drop(columns=['Unnamed: 0', 'uid', 'is_scammer'])
```

```python
y_test
```

#### Convert to Tensor

```python
x_test, y_test = tf.convert_to_tensor(x_test, dtype=tf.float32), tf.convert_to_tensor(y_test, dtype=tf.float32)
```

```python
model = tf.keras.models.load_model("saved_models/log_reg_keras_source_drop.h5", compile=False)
```

```python
model.summary()
```

```python
y_result = model.predict(x_test)
```

```python
len(y_result)
```

```python
pred_correct = 0
pred_false = 0
for i in range (0, len(y_result)):
    if y_result[i] >= 0.5:
        if (y_test[i] == 1):
            pred_correct += 1
        else:
            pred_false += 1
    else:
        if (y_test[i] == 0):
            pred_correct += 1
        else:
            pred_false += 1
```

```python
pred_correct
```

```python
pred_false
```

```python
# Accuracy
pred_correct/(pred_correct+pred_false)
```

```python
print("Accuracy: ", (TP+TN)/(TP+TN+FP+FN))
print("Precision: ", TP/(TP+FN))
print("Recall: ", TN/(TN+FP))
```
