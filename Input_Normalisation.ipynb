#### Import Libraries

```python
import pandas as pd
import numpy as np
from datetime import date
```

#### Load CSV File 

```python
df = pd.read_csv("user_data_test.csv",  sep=';')
```

```python
df.head()
```

#### Start Cleaning Here 

```python
df1 = df.copy()
```

```python
# Drop 'uid' and 'id' columns
df1 = df.drop(columns = ['uid', 'Unnamed: 0', 'is_scammer'])
df1.head()
```

```python
# Fill NULL data in column 8 - 42 (numeric data)
for i in range (8, 43):
    df1[df1.columns[i]] = df1[df1.columns[i]].fillna(0)
    
# Fill NULL data in date and time format columns
df1['trx_date'].fillna("11/01/2020") #18272
df1['report_date'].fillna("7/01/2020") #18268
df1['registereddate'].fillna("22/10/2013") #16000
df1['birthday'].fillna("7/12/1928") #-15000
```

```python
(pd.to_datetime("7/12/1928", format="%d/%m/%Y") - np.datetime64(0, 'Y')).days
```

```python
# Datetime-format columns change to integer
for i in range (1, 5):
    df1[df1.columns[i]] = pd.to_datetime(df1[df1.columns[i]])
    df1[df1.columns[i]] = (df1[df1.columns[i]] - np.datetime64(0, Y).dt.days)

df1['trx_date'] = pd.to_datetime(df1['trx_date'])
df1['trx_date'] = (df1['trx_date'] - np.datetime64(0, 'Y')).dt.days

df1['report_date'] = pd.to_datetime(df1['report_date'])
df1['report_date'] = (df1['report_date'] - np.datetime64(0, 'Y')).dt.days

df1['registereddate'] = pd.to_datetime(df1['registereddate'])
df1['registereddate'] = (df1['registereddate'] - np.datetime64(0, 'Y')).dt.days

df1['birthday'] = pd.to_datetime(df1['birthday'])
df1['birthday'] = (df1['birthday'] - np.datetime64(0, 'Y')).dt.days
```
